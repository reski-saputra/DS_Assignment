## text preprocessing

```python
#lower text
text = text.lower()
```

```python
#remove punctuation
text = re.sub(r'[^\w\s]',' ',text)
text
```

```python
def count(str):
    count=0
    words = str.split()
    for word in words:
        count+=1
    return count

def word_count(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
        
    return counts

def word_freq(str):
    counts = dict()
    words = str.split()

    for word in words:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
        
    return len(counts)
```

## a. What is the probability of the word “data” occurring in each line ?

```python
#number word data
word_data = word_count(text)
word_data['data']
```

```python
#probability of the word “data” occurring in each line
prob_word_data = word_data['data']/number_line
prob_word_data
```
